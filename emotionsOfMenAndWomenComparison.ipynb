```python
import gdown


url = 'https://drive.google.com/file/d/1xIkSZmk-qCamceXSHFkz_f_eMmwBEnqT/view?usp=drive_link'
file_id = url.split('/')[-2]
gdown_url = f'https://drive.google.com/uc?id={file_id}'
data = pd.read_csv(gdown_url)
print(data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Function to create a pie chart for a specific year
def create_pie_chart(year):
    # Filter the data for the given year
    selected_year_data = data[data['year'] == year]

    # Get the emotions percentages for the selected year
    emotions = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise']
    percentages = selected_year_data[emotions].values[0]

    # Create a pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(percentages, labels=emotions, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
    plt.axis('equal')
    plt.title(f'Emotions Distribution for Year {year}')
    plt.show()

# Choose the year for which you want to create the pie chart
selected_year = 2017  # You can change this to any other year from 2008 to 2017

# Create the pie chart for the selected year
create_pie_chart(selected_year)
```

```python
url = 'https://drive.google.com/file/d/1nN9Be_FgH0wlxp-xvtWXFF5oRITUp8zK/view?usp=sharing'
file_id = url.split('/')[-2]
gdown_url = f'https://drive.google.com/uc?id={file_id}'
data2 = pd.read_csv(gdown_url)
print(data2.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Function to create pie charts for a specific year
def create_pie_charts(year):
    # Filter the data for the given year
    selected_year_data = data2[data2['year'] == year]

    # Emotions and their corresponding labels for men and women
    emotions = ['angry', 'happy', 'sad']
    labels = ['Angry', 'Happy', 'Sad']

    # Get the emotions percentages for men and women for the selected year
    men_percentages = selected_year_data[['man_angry', 'man_happy', 'man_sad']].values[0]
    women_percentages = selected_year_data[['woman_angry', 'woman_happy', 'woman_sad']].values[0]

    # Create two pie charts, one for men and another for women
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # Pie chart for men
    axes[0].pie(men_percentages, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
    axes[0].axis('equal')
    axes[0].set_title(f'Emotions Distribution for Men in Year {year}')

    # Pie chart for women
    axes[1].pie(women_percentages, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
    axes[1].axis('equal')
    axes[1].set_title(f'Emotions Distribution for Women in Year {year}')

    plt.show()

# Choose the year for which you want to create the pie charts
selected_year = 2017  # You can change this to any other year from 2008 to 2017

# Create the pie charts for the selected year
create_pie_charts(selected_year)
```
